```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None

#node class does not link the nodes yet, just creates/generates the nodes
n1 = Node('Apple') #creates class object Node
n2 = Node('Banana') #creates


#Tail is always the "front" of the line!
class LinkedList: #linked list class is what makes nodes be in a list
    def __init__(self):
        self.head = None
        self.tail = None

    #enqueue is insert head
    def insert_head(self, value):
        new_node = Node(value)

        if self.head is None:      # list empty
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    #dequeue is delete tail
    def delete_tail(self): #removes the tail node and returns the value of the tail
        if self.tail is None:      # list empty
            return None
        #before we delete the tail, we grab the value to return it
        removed_value = self.tail.value

        if self.head == self.tail:  # only one node
            self.head = None
            self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None

        return removed_value

    def print_all(self):
        print('print_all')
        current = self.head
        while current:
            print(current.value)
            current = current.next

    def peek(self): #cleaned up to guard against an empty queue
      if self.tail is None:
        return None
      return self.tail.value



# Example usage
ll = LinkedList()

ll.insert_head("apple")
ll.insert_head("banana")
ll.insert_head("coconut")
ll.print_all()

print(ll.delete_tail())  # 10
print(ll.delete_tail())  # 20
```

1) Below is the Queue class I wrote. I made some tweaks to the existing Linked List class, mainly with peek(). I added an if statement to make sure the queue wasn't empty.

```python
class Queue:
  def __init__(self):
    self.internal_ll = LinkedList() #creating new LL variable that is internal
    #to linked list class
    self.n = 0

  def enqueue(self, value):
    self.internal_ll.insert_head(value)
    self.n += 1

  def dequeue(self):
    #only reduce counter if something was actually removed.
    removed = self.internal_ll.delete_tail() #this one doesn't need parameter
    if removed is not None:
      self.n -= 1
    return removed

  def peek(self):
    return self.internal_ll.peek()

  def num(self): #returns the number of nodes in the queue
    return self.n

```
