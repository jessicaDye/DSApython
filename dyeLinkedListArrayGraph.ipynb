## Definitions

**Linked Lists** - A collection of nodes, each node containing data. Can be singly-linked or doubly-linked; singly-linked goes in one direction, while doubly-linked has a pointer pointing both at previous and next node. End of linked list points at null or None. First node in linked list is the Head, last is the Taill.

**Array** - A collection of nodes stored in memory locations, with indexes. First index is always 0.

**Graph** - A data structure that is made of nodes referred to as vertices, and connections between nodes referred to as edges. Can be directed or undirected, can be weighted or unweighted.

**Binary Tree** - This is a Tree data structure, where the collection of nodes has a root/parent (first node), and no more than two child nodes. Child nodes are on the left and right.

**Queue** - A linear data structure that uses FIFO (First in, First out)

**Stack** - A linear data structure that uses LIFO (Last in, First Out) (Also can be remembered *First Hired, Last Fired*)

**Hash Table** - A data structure that maps keys to values, using a Hash function for fast access.



```python
# Doubly Linked List
# This doubly linked list I built functions like a queue, using the FIFO method
# It inserts at elements at tail, and deletes at head.
# It is a doubly-linked list making it structurally different than a queue
# These two member functions/methods are O(1), Constant time making them quick
# These two methods can be compared to enqueue() and dequeue()
# Being doubly-linked instead of singly gives slighly more flexibility
# than a queue due to having next and previous ptrs.


#Node class
class Node:
  def __init__(self, data): #__init__ is Constructor in Python
    self.data = data
    self.next = None
    self.prev = None

#Doubly Linked List class
class DoublyLinkedList:
    def __init__(self):
      self.head = None
      self.tail = None

    def isEmpty(self):
      if self.head == None:
        return True
      else:
        return False



    #insert at tail member function/method
    #insertAtTail is O(1) Constant time
    def insertAtTail(self, node):
      if self.head is None:
        self.head = node
        self.tail = node
      else:
        self.tail.next = node
        node.prev = self.tail
        self.tail = node

    def deleteAtHead(self):
      if self.head == None:
        return None

      removed = self.head

      if self.head == self.tail:
        self.head = None
        self.tail = None
      else:
        self.head = self.head.next
        self.head.prev = None

      return removed

    #print function, prints head to tail data
    def printHeadToTailData(self):
      current = self.head

      while current is not None:
        print(current.data)
        current = current.next


```

```python
#Doubly Linked List, implementing Node and DoublyLinkedList classes
dll = DoublyLinkedList()

dll.insertAtTail(Node(6))
dll.insertAtTail(Node(7))
dll.insertAtTail(Node(8))
dll.insertAtTail(Node(10))

#Initial List
dll.printHeadToTailData()

removed = dll.deleteAtHead()
print("Removed: ", removed.data)

print("After deleting Head: ")
dll.printHeadToTailData()

#Deleting all elements
dll.deleteAtHead()
dll.deleteAtHead()
dll.deleteAtHead()

#Testing if all elements are removed:
dll.printHeadToTailData()
```

```python
#Array Structure : makes a fixed size Array
class Array:
  def __init__(self, n):
    self.size = n
    self.data = [None] * n

  def set(self, i, value):
    self.data[i] = value

  def get(self, i):
    return self.data[i]
```

```python
#Testing out fixed size array
myArray = Array(5)

myArray.set(0, "A")
myArray.set(1, "B")
myArray.set(2, "C")

print(myArray.get(0))
print(myArray.get(1))
print(myArray.get(2))

#Checking the empty slots
print(myArray.get(3))
print(myArray.get(4))
```

##Time Complexity in Array

Difference between list function .append() - append() can be O(1), but other times be O(n) because of resizing. Again, list is dyanmic meaning it can be resized where an array is a fixed structure.



```python
#Graph - undirected, adjacency matrix implementation

class Graph:
  def __init__(self, lables):
    self.lables = lables
    self.size = len(lables)

    #Creates matrix as part of constructor, filled w/ 0's
    self.matrix = []
    for i in range(self.size):
      row = []
      for j in range(self.size):
        row.append(0)
      self.matrix.append(row)

  #Helper function to get the index
  def get_index(self, lables):
    for i in range(self.size):
      if self.lables[i] == lables:
        return i
    return -1 #-1 is not found

  #add undirected edge
  def add_edge(self, lableA, lableB):
    i = self.get_index(lableA)
    j = self.get_index(lableB)

    if i == -1 or j == -1:
      return

    self.matrix[i][j] = 1
    self.matrix[j][i] = 1

  def has_connection(self, vertexA, vertexB):
    i = self.get_index(vertexA)
    j = self.get_index(vertexB)

    if i == -1 or j == -1:
      return False

    return self.matrix[i][j] == 1




```
