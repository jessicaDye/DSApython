A **Queue** data structure is also called a **First In First Out (FIFO).**  

A **Queue** data structure has the following functions:

* enqueue(new_element) - adds new element to the "back" of the queue.
* dequeue() - removes an element from the "front" of the queue.
* peak() - returns the element at the "front".
* num() - returns a count of the number of elements still in the queue.



A python list allows a simple implementation of a Queue, with the undesireable trait that the dequeue function is O(n) instead of O(1).

```python
# Implement a queue with a list

class Queue:
  def __init__(self):
    self.q = []
  def enqueue(self, new_element):
    self.q.append(new_element)
  def dequeue(self):
    self.q.pop(0)
  def peek(self):
    return self.q[0]
  def num(self):
    return len(self.q)
```

```python
# TEST CASE

q = Queue()
q.enqueue("Cue")
q.enqueue("Ayo")
q.enqueue("Damian")
print('peek should be Cue: ', q.peek()) # should be "Cue"

q.dequeue() # removes Cue
print("should be: Ayo", q.peek()) # should be "Ayo"

print("should be 2: ", q.num()) # should be 2

q.dequeue() # removes "Ayo"

print("should be Damian: ", q.peek())  # should be Damian

```
